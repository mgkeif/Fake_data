###This program generates a PDF target for color matching of brand "critical" colors (spot colors) on a digital press.

To create your custom target, start the program by clicking on the arrow left of "Malcolms Target Generator". Input your filename, as well as your CMYK values. You may need to scroll down to the bottom of the code section to see the input fields if you have expanded the code section.

```python
#@title Malcolms Target Generator
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Feb 25 16:27:55 2021

@author: Malcolm Keif, Cal Poly
"""
import sys
!{sys.executable} -m pip install reportlab

from reportlab.pdfgen import canvas
from reportlab.lib.colors import PCMYKColor
from reportlab.lib.units import inch
from reportlab.lib.pagesizes import letter

        
# class Page(object):
#     def __init__(self, pagesize): 
#         self.pagesize=letter
# def pagesize(canvas):

#Input name of the file. 
PMS = input ('Input your filename:')
filename = PMS + ".pdf"
ci = round(float (input ('Input Cyan value from 0-100%:')),1)
mi = round(float (input ('Input Magenta value from 0-100%:')),1)
yi = round(float (input ('Input Yellow value from 0-100%:')),1)
ki = round(float (input ('Input Black value from 0-100%:')),1)

#ci = 12
#mi = 84
#yi = 80
#ki = 0

#myCanvas = canvas.Canvas("test5.pdf", pagesize=letter, bottomup=0) 
myCanvas = canvas.Canvas(filename, pagesize=letter, bottomup=0) 
cwidth, cheight = letter

# starty = cheight/16
# startx = cwidth/16

R1 = cheight*.10
R2 = cheight*.42
R3 = cheight*.74
Soff = -47

gray = PCMYKColor(0,0,0,50)
black = PCMYKColor(0,0,0,100)
white = PCMYKColor(0,0,0,0)
g7 = PCMYKColor(50,40,40,0)
inc1 = .02
# inc2 = .01
# incp = .005

spotsize = .45

w = h = inch * spotsize
# Adds 1/2 point between patches
dy = w + float(0.5)
dx = h + float(0.5)
   

pcmyk = ((ci),(mi),(yi),(ki))
c,m,y1,k = pcmyk

def Center (canvas):
    canvas.setStrokeColor(black)
    canvas.setLineWidth(1)
    canvas.rect(inch*.25,inch*3.75,inch*8,inch*3.5,fill=0,stroke=1)

def Sep (canvas,startx,starty,Kval):
    canvas.setFillColor(gray)
    canvas.setStrokeColor(g7)
    canvas.setLineWidth(4)
    canvas.rect(startx+inch*2.0,starty,inch*3.5,20,fill=1,stroke=1)
    canvas.setFillColor(white)
    canvas.setFont("Helvetica",12)
    canvas.drawCentredString(cwidth/2,starty+15,str(Kval))
    
def colorsCMYK(canvas, startx, starty, ci, mi, yi, ki, static):
    # Defines a black and white in CMYK ; PCMYKColor uses integer values (decimals = %)        
    # size = float (input('Input single patch size in inches (0.45 recommended)')0
    
    #Sets the starting point on the page for the rectangles in inches
    y = starty; x = startx    
    
    cp = ci / 100.0
    mp = mi / 100.0
    yp = yi / 100.0
    kp = ki / 100.0
    
    if(static == "Yellow"):
    
        # Set while offset for x and y
        wx = wy = 0
        cp = cp - 2 * inc1
        mp = mp + 2 * inc1
        yp = yp
        
        # Rows while statement
        while wy < 5:
            canvas.setFillColorCMYK(cp,mp,yp,kp)
            canvas.setStrokeColor(white)
            canvas.setLineWidth(0.5)
            # Columns while statement
            while wx < 5:
                canvas.setFillColorCMYK(cp,mp,yp,kp)
                canvas.rect(x, y, w, h, fill=1, stroke=0)
                x = x + dx
                cp = cp + inc1
                wx = wx + 1
            cp = cp - 0.10
            wx = 0
            x = x - 5 * dx
            y = y + dy
            mp = mp - inc1
            wy = wy + 1 
        canvas.setFillColorCMYK(cp,mp,yp,kp)
        canvas.setStrokeColor(black)
        canvas.setLineWidth(0.5)
        x = x + 2*dx
        y = y - 3*dy
        cp = cp - 2 * inc1
        mp = mp + 2 * inc1
        canvas.rect(x, y, w, h, fill=0, stroke=1)
        
        canvas.setFillColor(black)
        canvas.setFont("Helvetica",11)
        canvas.drawCentredString(x + dx*0.5 , y - dy*2.25, "C%s  M%s  Y%s  K%s"%pcmyk)
                                 # ci+"%, "mi+"%, "yi"%, "ki"%")
        canvas.setFont("Helvetica",7)
        canvas.drawCentredString(x + dx*0.5, y + dy*3.25,"-"+str(2*inc1)+"C <---------------------------------------> +"+str(2*inc1)+"C")
        canvas.rotate(270)
        canvas.drawCentredString(y - dx*24.7, x - dy*2.12,"-"+str(2*inc1)+"M <---------------------------------------> +"+str(2*inc1)+"M")
        canvas.rotate(-270)
        
    elif (static == "Magenta"):
          # Set while offset for x and y
        wx = wy = 0
        cp = cp - 2 * inc1
        mp = mp
        yp = yp + 2 * inc1
        
        # Rows while statement
        while wy < 5:
            canvas.setFillColorCMYK(cp,mp,yp,kp)
            canvas.setStrokeColor(white)
            canvas.setLineWidth(0.5)
            # Columns while statement
            while wx < 5:
                canvas.setFillColorCMYK(cp,mp,yp,kp)
                canvas.rect(x, y, w, h, fill=1, stroke=0)
                x = x + dx
                cp = cp + inc1
                wx = wx + 1
            cp = cp - 0.10
            wx = 0
            x = x - 5 * dx
            y = y + dy
            yp = yp - inc1
            wy = wy + 1 
        canvas.setFillColorCMYK(cp,mp,yp,kp)
        canvas.setStrokeColor(black)
        canvas.setLineWidth(0.5)
        x = x + 2*dx
        y = y - 3*dy
        cp = cp - 2 * inc1
        yp = yp + 2 * inc1
        canvas.rect(x, y, w, h, fill=0, stroke=1)
        
        canvas.setFillColor(black)
        canvas.setFont("Helvetica",11)
        canvas.drawCentredString(x + dx*0.5 , y - dy*2.25, "C%s  M%s  Y%s  K%s"%pcmyk)
                                  # ci+"%, "mi+"%, "yi"%, "ki"%")
        canvas.setFont("Helvetica",7)
        canvas.drawCentredString(x + dx*0.5, y + dy*3.25,"-"+str(2*inc1)+"C <---------------------------------------> +"+str(2*inc1)+"C")
        canvas.rotate(270)
        canvas.drawCentredString(y - dx*24.7, x - dy*2.12,"-"+str(2*inc1)+"Y <---------------------------------------> +"+str(2*inc1)+"Y")
        canvas.rotate(-270)
    
    elif (static == "Cyan"):
          # Set while offset for x and y
        wx = wy = 0
        cp = cp
        mp = mp + 2 * inc1
        yp = yp - 2 * inc1
        
        # Rows while statement
        while wy < 5:
            canvas.setFillColorCMYK(cp,mp,yp,kp)
            canvas.setStrokeColor(white)
            canvas.setLineWidth(0.5)
            # Columns while statement
            while wx < 5:
                canvas.setFillColorCMYK(cp,mp,yp,kp)
                canvas.rect(x, y, w, h, fill=1, stroke=0)
                x = x + dx
                yp = yp + inc1
                wx = wx + 1
            yp = yp - 0.10
            wx = 0
            x = x - 5 * dx
            y = y + dy
            mp = mp - inc1
            wy = wy + 1 
        canvas.setFillColorCMYK(cp,mp,yp,kp)
        canvas.setStrokeColor(black)
        canvas.setLineWidth(0.5)
        x = x + 2*dx
        y = y - 3*dy
        yp = yp - 2 * inc1
        mp = mp + 2 * inc1
        canvas.rect(x, y, w, h, fill=0, stroke=1)
        
        canvas.setFillColor(black)
        canvas.setFont("Helvetica",11)
        canvas.drawCentredString(x + dx*0.5 , y - dy*2.25, "C%s  M%s  Y%s  K%s"%pcmyk)
                                  # ci+"%, "mi+"%, "yi"%, "ki"%")
        canvas.setFont("Helvetica",7)
        canvas.drawCentredString(x + dx*0.5, y + dy*3.25,"-"+str(2*inc1)+"Y <---------------------------------------> +"+str(2*inc1)+"Y")
        canvas.rotate(270)
        canvas.drawCentredString(y - dx*24.7, x - dy*2.12,"-"+str(2*inc1)+"M <---------------------------------------> +"+str(2*inc1)+"M")
        canvas.rotate(-270)
    
    
colorsCMYK(myCanvas,cwidth*0.0625,R1,ci,mi,yi,ki+2.0,"Yellow")
colorsCMYK(myCanvas,cwidth*0.368,R1,ci,mi,yi,ki+2.0,"Magenta")
colorsCMYK(myCanvas,cwidth*0.675,R1,ci,mi,yi,ki+2.0,"Cyan")
Sep(myCanvas,cwidth/16,R1+Soff,"K +2.0%")

colorsCMYK(myCanvas,cwidth*0.0625,R2,ci,mi,yi,ki,"Yellow")
colorsCMYK(myCanvas,cwidth*0.368,R2,ci,mi,yi,ki,"Magenta")
colorsCMYK(myCanvas,cwidth*0.675,R2,ci,mi,yi,ki,"Cyan")
Sep(myCanvas,cwidth/16,R2+Soff,"K Static")
Center(myCanvas)

colorsCMYK(myCanvas,cwidth*0.0625,R3,ci,mi,yi,ki-2.0,"Yellow")
colorsCMYK(myCanvas,cwidth*0.368,R3,ci,mi,yi,ki-2.0,"Magenta")
colorsCMYK(myCanvas,cwidth*0.675,R3,ci,mi,yi,ki-2.0,"Cyan")
Sep(myCanvas,cwidth/16,R3+Soff,"K -2.0%")

myCanvas.showPage()
myCanvas.save()


```
